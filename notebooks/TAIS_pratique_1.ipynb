```python
import requests
import csv
import json
import time


base_url = "https://api-adresse.data.gouv.fr/search/?"
out = open("3_DataGouv_geoloc_prop.csv", "w")
out_writer = csv.writer(out, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
source_prop = "https://api-adresse.data.gouv.fr"
								
with open('0_listeAdressesDedoubl_AZ.csv') as csv_file:
	csv_reader = csv.reader(csv_file, delimiter=',')
	line_count = 0
	for row in csv_reader:
		if line_count == 0:
			out_writer.writerow(['index*','adresse', 'num', 'proprio[global]', 'collectif', 'sexe', 'statut_femme', 'proprio_titre*', 'proprio_particule*', 'proprio_name*', 'proprio_ad', 'proprio_numAd', 'commune_propio', 'proprio_region*', 'Pays', 'same_place*', 'commentaire', 'x_prop', 'y_prop', 'postcode_en2020_prop', 'source_prop', 'url_prop', 'score_prop', 'label_adresse'])
			line_count += 1
		else:
			line_count += 1
			adres = row[2] + " " + row[1]
			params = {"q": adres.replace(" ", "+"), "city": "paris", "limit": "1000"}
			#print("adresse: "+row[2]+" "+row[1])
			url_prop = base_url + "q" + "=" + adres.replace(" ", "+") + "&" + "city=paris" +  "&" + "limit=1000"
			#print("requete: "+url)
			r = requests.get(base_url, params)
			is_paris = False
			if r.status_code == 200:
				if (r.json()):
					if (len(r.json()['features']) > 0):
						for resultat in r.json()['features']:
							if resultat['properties']['city'] == "Paris":
								is_paris = True
								print(resultat)
								x_prop = resultat['geometry']['coordinates'][0]
								y_prop = resultat['geometry']['coordinates'][1]
								postc_prop = resultat['properties']['postcode']
								score_prop = resultat['properties']['score']
								label_add = resultat['properties']['label']
								out_writer.writerow([row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], row[11], row[12], row[13], row[14], row[15], row[16], x_prop, y_prop, postc_prop, source_prop, url_prop, score_prop, label_add])
						if is_paris == False:
							print("ADRESSE A VERIFIER:" + row[2] + " " + row[1])
							out_writer.writerow([row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], row[11], row[12], row[13], row[14], row[15], row[16], "", "", "", "", url_prop, "", ""])
					else:
						print("ADRESSE A VERIFIER:" + row[2] + " " + row[1])
						out_writer.writerow([row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], row[11], row[12], row[13], row[14], row[15], row[16], "", "", "", "", url_prop, "", ""])
				else:
					print("ADRESSE A VERIFIER:" + row[2] + " " + row[1])
					out_writer.writerow([row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], row[11], row[12], row[13], row[14], row[15], row[16], "", "", "", "", url_prop, "", ""])
			if (line_count % 500 == 0):
				time.sleep(5)
out.close()
```
